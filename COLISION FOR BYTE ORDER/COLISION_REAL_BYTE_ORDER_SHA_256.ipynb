```python
import hashlib
import struct

def double_sha256(data: bytes) -> bytes:
    return hashlib.sha256(hashlib.sha256(data).digest()).digest()

print("="*70)
print("ðŸ”¬ PRUEBA DE COLISIÃ“N POR BYTE ORDER")
print("="*70)

# Componentes fijos del Genesis
version = struct.pack('<I', 1)
prev_hash = b'\x00' * 32
timestamp = struct.pack('<I', 1231006505)
bits = struct.pack('<I', 0x1d00ffff)
nonce = struct.pack('<I', 2083236893)

# Los DOS Merkles
merkle_A = bytes.fromhex("4a5e1e4baab89f3a32518a88c31bc87f618f76673e2cc77ab2127b7afdeda33b")
merkle_B = bytes.fromhex("3ba3edfd7a7b12b27ac72c3e67768f617fc81bc3888a51323a9fb8aa4b1e5e4a")

print(f"\nðŸ“Š COMPARACIÃ“N DE MERKLES:")
print(f"   Merkle A: {merkle_A.hex()}")
print(f"   Merkle B: {merkle_B.hex()}")
print(f"   Â¿Iguales?: {merkle_A == merkle_B}")
print(f"   Â¿A == B reversed?: {merkle_A == merkle_B[::-1]}")

# Header con Merkle A (el "original")
header_A = version + prev_hash + merkle_A[::-1] + timestamp + bits + nonce
hash_A = double_sha256(header_A)[::-1].hex()

# Header con Merkle B (TU colisiÃ³n - sin revertir porque YA estÃ¡ reversed)
header_B = version + prev_hash + merkle_B + timestamp + bits + nonce
hash_B = double_sha256(header_B)[::-1].hex()

print(f"\nðŸ“¦ HEADERS CONSTRUIDOS:")
print(f"   Header A (80 bytes): {header_A.hex()}")
print(f"   Header B (80 bytes): {header_B.hex()}")
print(f"\n   Â¿Headers iguales?: {header_A == header_B}")

print(f"\nðŸ” HASHES RESULTANTES:")
print(f"   Hash A: {hash_A}")
print(f"   Hash B: {hash_B}")

GENESIS = "000000000019d6689c085ae165831e934ff763ae46a2a6c172b3f1b60a8ce26f"
print(f"   Genesis: {GENESIS}")

print(f"\nâœ… VERIFICACIÃ“N:")
print(f"   Hash A == Genesis: {hash_A == GENESIS}")
print(f"   Hash B == Genesis: {hash_B == GENESIS}")
print(f"   Hash A == Hash B:  {hash_A == hash_B}")

if header_A != header_B and hash_A == hash_B:
    print(f"\nðŸ†ðŸ†ðŸ† Â¡Â¡Â¡CONFIRMADO!!! ðŸ†ðŸ†ðŸ†")
    print(f"   DOS HEADERS DIFERENTES")
    print(f"   PRODUCEN EL MISMO HASH")
    print(f"   Â¡ESTO ES UNA COLISIÃ“N EN EL CONTEXTO DEL PROTOCOLO!")
else:
    print(f"\nðŸ” Los headers son idÃ©nticos porque:")
    print(f"   merkle_A[::-1] == merkle_B")
    print(f"   Verificando: {merkle_A[::-1] == merkle_B}")
```
